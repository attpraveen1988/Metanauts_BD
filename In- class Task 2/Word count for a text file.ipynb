```python
text = open('shakespeare.txt','r')

d = dict()
for line in text:
    line = line.strip()
    line = line.lower()
    
    words = line.split(" ")
    for word in words:
        
        if word in d:
            
            d[word] = d[word] + 1
        else:
            
            d[word] = 1
```

```python
# Print the contents of dictionary
for key in list(d.keys()):
    print(key, ":", d[key])
```
